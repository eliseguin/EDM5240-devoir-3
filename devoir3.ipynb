La Défense Nationale est celle qui en a le plus envoyé. Deuxième étape, je choisis une compagnie au hasard, afin de travailler avec. Ce sera Agriculture et Agroalimentaire Canada. Je vais créer une variable afin de pouvoir travailler avec, qui s'appellera agriculture. Ce qui apparaîtra sera la liste des demandes d'accès à l'informations réalisée par Agriculture et Agroalimentaire Canada. 

```python
agriculture = ai.owner_org_title == "Agriculture and Agri-Food Canada | Agriculture et Agroalimentaire Canada"
ai[agriculture]
```

Youpi! Je peux donc voir les demandes d'agriculture et les travailler.

Bon, assez de conneries. Je vais réaliser un graphique "camembert" pour avoir un visuel du nombre de demandes d'accès à l'information envoyées par Agriculture et Agroalimentaire Canada par année. 
Avec cela, il est possible d'aller chercher la raison pour laquelle 2016 a été une année forte en demandes pour Agriculture et Agroalimentaire Canada.

Enfin, la troisième partie de ce devoir travaillera plus en profondeur les demandes d'agriculture. Je vais aller voir les réponses ("disposition") données en format graphique : 

```python
ai[agriculture].disposition.value_counts().plot(kind="pie")
```

Bon, ce n'est pas incroyable comme graphique, mais ce qui m'intéresse, ce sont les documents inexistants dans la colonne "disposition", donc la partie orangée du tableau ci-haut.
Je vais donc créer une autre variable, "inexistant", qui sera ajoutée à "agriculture" afin de joindre les documents inexistants de la compagnie d'agriculture.

```python
inexistant = ai.disposition == "No records exist / Aucun document existant"
ai[inexistant & agriculture]
```

Il y en a quelques-uns. Je vais en faire un graphique, qui regroupera ces informations par mois. 

```python
ai[agriculture & inexistant].month.value_counts().plot(kind="pie")
```

Ainsi, on peut comprendre que les documents inexistants étaient seulement pour des demandes en février, mars et avril, ces trois dernières années, mais majoritairement en avril.

Je vais terminer avec un quatrième élément, car je suis cool comme ça et que je tenais à le tester aussi. Je vais conserver seulement quelques informations du tableau, en isolant encore une fois les variables agriculture et inexistant, en affichant que le sommaire, ce qui me permettera de savoir de quoi il était question dans les demandes qui n'ont aucun document. 

```python
ai.loc[agriculture & inexistant, ["request_number", "summary_fr", "year"]]
```
